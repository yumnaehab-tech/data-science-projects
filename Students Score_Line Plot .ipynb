**By Yumna Ehab**

This project visualizes the average scores of students in different subjects using Matplotlib.  
The dataset is from Kaggle: [Students Performance in Exams](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)  

**Objective:**  
- Calculate average scores for each subject.  
- Plot a line chart to show the trend.  
- Extract insights from the visualization.



```python
import zipfile
import io

zip_file = "StudentsPerformance.zip"
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
columns_to_use = ['math score', 'reading score', 'writing score']
df_scores = df[columns_to_use]
df_scores.head()
```

```python
average_scores = df_scores.mean()
print(average_scores)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(average_scores.index, average_scores.values, marker='o', linestyle='-', color='b')

plt.title('Average Student Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

#   
plt.plot(average_scores.index, average_scores.values, marker='o', linestyle='-', color='green')

#  
plt.title('Average Student Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(True)

#  annotation  
highest_subject = average_scores.idxmax()
highest_score = average_scores.max()
plt.annotate('Highest score', xy=(highest_subject, highest_score),
             xytext=(1, 90), arrowprops=dict(facecolor='black', shrink=0.05))

#  
plt.show()
```
