# Distribution of Math Scores (Histogram)
**By Yumna Ehab**

This project visualizes how students' math scores are distributed  
using a histogram to understand performance spread and concentration.


```python
from google.colab import files
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

uploaded = files.upload()

zip_file = "Students Performance in Exams (1).zip"
with zipfile.ZipFile(zip_file, 'r') as z:
    csv_file = [f for f in z.namelist() if f.endswith('.csv')][0]
    z.extractall()

df = pd.read_csv(csv_file)

scores = pd.to_numeric(df['math score'], errors='coerce').dropna()

sns.set(style="whitegrid")
plt.figure(figsize=(9,5))
sns.histplot(scores, bins=12, kde=True, stat='count', edgecolor='black', alpha=0.7)
plt.title('Distribution of Math Scores with KDE', fontsize=14, weight='bold')
plt.xlabel('Math Score', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xlim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.savefig('math_score_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
